```python
# 读取celltype信息
anno_cell = pd.read_csv("/home/zhangyinan/TXL_new/4_cell2location/Integrate/Control5_Model2_Treatment3/result/celltype/celltype_first.csv", index_col=0)
anno_cell.head()

# 读取subtype信息
anno_sub = pd.read_csv("/home/zhangyinan/TXL_new/4_cell2location/Integrate/Control5_Model2_Treatment3/result/subtype_202507/subtype_first.csv", index_col=0)
anno_sub.head()

# 读取singleR数据
anno_sin = pd.read_csv("/home/zhangyinan/TXL_new/2_annotation/Control5_Model2_Treatment3/result/anno_single_c5_m2_t3.csv", index_col = 0)
anno_sin.head()
```

```python
## 新建一个注释文件
# 创建一个新的 DataFrame anno，并以 anno_sin 的索引为索引
anno = pd.DataFrame(index=anno_cell.index)

# 添加列
anno["cell_type_c2l"] = anno_cell.loc[anno.index, "cell_type"]
anno["sub_type_c2l"] = anno_sub.loc[anno.index, "cell_type"]
anno["singleR"] = anno_sin.loc[anno.index, "labels"]
```

```python
anno.to_csv('/home/zhangyinan/TXL_new/4_cell2location/Integrate/Control5_Model2_Treatment3/result/annotation_202507.csv', index = True)
```
